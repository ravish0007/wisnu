## Finding the number of <big><i><font color="red">Nan</font></i></big> in the dataset, by adding number of Nans in columns and rows
  * Using isnull() gives logical values of presence of data, True corresponds to 1 and False to 0
 * Using sum() once gives number of Nans in columns, another sum() adds this vector

```python
x = datasheet.isnull().sum().sum()
print("Total missing values (Nan) = ",x)
```

## Finding the number of valid data cells

```python
y = datasheet.count().sum()
print("Total amount of datapoint cells = ",y)
```

## Finding percentage of Missing values

```python
percentage = x/(y+x)
print("Percentage of missing values = ", round(percentage*100,3))
```

## <font color="red"> 1) Cleaning the data</font>, by getting rid of rows with Nan values

* Note: All the columns will be dropped if you set axis to 1

```python
data_missing_rows = datasheet.dropna(axis=0)
data_missing_rows
```

## Finding percentage of retrieved values from old dataset

```python
retrieved_percent = len(data_missing_rows)/len(datasheet)
print("Percentage of data retrieved = ",retrieved_percent)
```

## <font color="red">2) Cleaning the data</font>, by replacing Nan values with 0


```python
data_with_zeroes = datasheet.fillna(0)
data_with_zeroes.style
```

## <font color="red">3) Cleaning the data</font>, by replacing Nan values with previous values

* However, It doesnot replaces the values of the top row due to absence of previous row
* Can be done to columns as well, In this particular case it will be invalid

```python
data_with_previous_values  = datasheet.fillna(method='ffill', axis=0 )
data_with_previous_values.style
```

## <font color="red">4) Cleaning the data</font>, by replacing Nan values with next row's values

* However, It doesnot replaces the values of the bottom rows due to absence of next rows
* Can be done to columns as well, In this particular case it will be invalid

```python
data_with_previous_values = datasheet.fillna(method='backfill', axis=0 )
data_with_previous_values.style
```

## <font color="red">5) Cleaning the data</font>, by replacing Nan with values linear interpolated values
* However, It doesnot replaces the  some values because of absence of values required to calculate linear interpolation 


```python
data_linear_values = datasheet.interpolate(method='linear', axis=0 )
data_linear_values.style
```

```python
from numpy import nan
from pprint import pprint
import re



#Finding the unique Genre

genre = set(   i.upper() for i in datasheet['Genre'] )

genre =  {  i.replace(',','') for i in genre}

genre = { i.replace(' ', '') for i in genre}

genre = { i.rstrip().lstrip() for i in genre}

#processed genres

#pprint(genre)



# For each Genre assigning a dictionary of datapoints to store accumulated value 

average_dictionay = \
{ i:{'Duration':0, 'Rating':0, 'Voting':0, 'Viewer':0,  'Dlen':0, 'Rlen':0, 'Volen':0, 'Vilen':0 } for i in genre }



for g,d,r,vo,vi in zip(datasheet['Genre'],datasheet['Duration'],datasheet['Rating'],datasheet['Voting'],datasheet['Viewer']):
    
    g = g.upper().replace(',','').replace(' ','').rstrip().lstrip()
    if  d != nan:
        average_dictionay[g]['Duration'] = d 
        average_dictionay[g]['Dlen'] += 1
        
    if  r != nan:
        average_dictionay[g]['Rating'] = r 
        average_dictionay[g]['Rlen'] += 1
        
    if vo != nan:
        average_dictionay[g]['Voting'] = vo 
        average_dictionay[g]['Volen'] += 1       
        
    if vi != nan:
        average_dictionay[g]['Viewer'] = vi 
        average_dictionay[g]['Vilen'] += 1       

        
# the following data could be used instead of replacing zeroes or backward/forward values
# rlen,volen,vilen,dlen corresponds to length  of that specific genre

print("The Following data holds average values to specific genre and some nan values")

average_dictionay = pd.DataFrame(average_dictionay)

average_dictionay.style



```
